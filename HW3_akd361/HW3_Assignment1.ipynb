```python
from __future__ import print_function
__author__= 'akd'

import pylab as pl
import seaborn
from scipy.optimize import curve_fit, minimize
import matplotlib as plt

%pylab inline
```

```python
#distributions = ['chisq', 'lnorm', 'binomial', 'pois', 'gaus']
distributions = ['chisq']
```

```python
mymean = 100
```

```python
df = mymean
```

```python
md = {}
```

```python
md['chisq'] = np.random.chisquare(df, size=100)
```

```python
pl.hist(md['chisq'], bins = 100)
pl.ylabel('y')
pl.xlabel('x')
```

```python
print ("Chisq mean: %.2f, standard deviation: %.2f"%(md['chisq'].mean(), md['chisq'].std()))
```

```python
mysize = (2000 / (np.array(range(1, 100)))).astype(int)
print (mysize, mysize.shape)
```

```python
md['chisq'] = {} 
```

```python
for n in mysize:
    md['chisq'][n] = np.random.chisquare(df, size = n)
```

```python
md['chisq']['means'] = {}
axchisq_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['chisq']:
    if not type(nn) == str:
        md['chisq']['means'][nn] = md['chisq'][nn].mean()

        axchisq_mu_n.plot(nn, md['chisq']['means'][nn], 'o')
        axchisq_mu_n.set_xlabel('sample size', fontsize=18)
        axchisq_mu_n.set_ylabel('sample mean', fontsize=18)
        axchisq_mu_n.set_title('Chi squared', fontsize=18)
        axchisq_mu_n.plot([min(mysize), max(mysize)], [df, df], 'k')
```

```python
# Explanation: ghe larger your sample size, the closer the sample's mean is likely to
# be to the mean of the population.
```

```python
allmeansch = list(md['chisq']['means'].values())

pl.figure(figsize=(10, 10))
pl.hist(allmeans,bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('Number of samples', fontsize = 18)
```

POISSON DISTRIBUTION

```python
distributions = ['pois']

md['pois'] = np.random.poisson(lam=2, size=100)

pl.hist(md['pois'], bins = 30)
pl.ylabel('y')
pl.xlabel('x')

print ("Poisson mean: %.2f, standard deviation: %.2f"%(md['pois'].mean(), md['pois'].std()))

mysize = (2000.0 / (np.array(range(1, 100)))).astype(int)
print (mysize, mysize.shape)

md['pois'] = {}

for n in mysize:
    md['pois'][n] = np.random.poisson(100, size = n)
    
    md['pois']['means'] = {}
poisson = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['pois']:
    if not type(nn) == str:
        md['pois']['means'][nn] = md['pois'][nn].mean()
        poisson.plot(nn, md['pois']['means'][nn], 'o')
        poisson.set_xlabel('sample size', fontsize=18)
        poisson.set_ylabel('sample mean', fontsize=18)
        poisson.set_title('Poisson', fontsize=18)
        poisson.plot([min(mysize), max(mysize)], [10, 10], 'k')
        
allmeanspo = list(md['pois']['means'].values())

pl.figure(figsize=(10, 10))
pl.hist(allmeans,bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

BINOMIAL DISTRIBUTION

```python
distributions = ['binomial']

md['binomial'] = np.random.binomial(200, .5, size=100)

pl.hist(md['binomial'], bins = 30)
pl.ylabel('y')
pl.xlabel('x')

print ("Binomial mean: %.2f, standard deviation: %.2f"%(md['binomial'].mean(), md['binomial'].std()))

mysize = (2000 / (np.array(range(1, 100)))).astype(int)
print (mysize, mysize.shape)

md['binomial'] = {}

for n in mysize:
    md['binomial'][n] = np.random.binomial(200, .5, size = n)
    
    md['binomial']['means'] = {}
binomialdist = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['binomial']:
    if not type(nn) == str:
        md['binomial']['means'][nn] = md['binomial'][nn].mean()
        binomialdist.plot(nn, md['binomial']['means'][nn], 'o')
        binomialdist.set_xlabel('sample size', fontsize=18)
        binomialdist.set_ylabel('sample mean', fontsize=18)
        binomialdist.set_title('Binomial', fontsize=18)
        binomialdist.plot([min(mysize), max(mysize)], [50, 50], 'k')
        
allmeansbi = list(md['binomial']['means'].values())

pl.figure(figsize=(10, 10))
pl.hist(allmeans,bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

CAUCHY DISTRIBUTION

```python
distributions = ['cauchy']

md['cauchy'] = numpy.random.standard_cauchy(100)

pl.hist(md['cauchy'], bins = 30)
pl.ylabel('y')
pl.xlabel('x')

print ("Cauchy mean: %.2f, standard deviation: %.2f"%(md['cauchy'].mean(), md['cauchy'].std()))

mysize = (2000 / (np.array(range(1, 100)))).astype(int)
print (mysize, mysize.shape)

md['cauchy'] = {}

for n in mysize:
    md['cauchy'][n] = np.random.poisson(100, size = n)
    
    md['cauchy']['means'] = {}
caucheese = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['cauchy']:
    if not type(nn) == str:
        md['cauchy']['means'][nn] = md['cauchy'][nn].mean()
        caucheese.plot(nn, md['cauchy']['means'][nn], 'o')
        caucheese.set_xlabel('sample size', fontsize=18)
        caucheese.set_ylabel('sample mean', fontsize=18)
        caucheese.set_title('Cauchy', fontsize=18)
        caucheese.plot([min(mysize), max(mysize)], [2, 2], 'k')
        
allmeansca = list(md['cauchy']['means'].values())

ax = plt.figure(figsize=(10, 10))
pl.hist(allmeans,bins=30,alpha=.5)
#pl.hist(allmeansbi,bins=30,alpha=.5)
#pl.hist(allmeanspo,bins=30,alpha=.5)
#pl.hist(allmeansno,bins=30,alpha=.5)
#pl.hist(allmeansch,bins=30,alpha=.5)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```
